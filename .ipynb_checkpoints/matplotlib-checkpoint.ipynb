```python
x = np.linspace(-10, 10, 40)
y = x**2

x_points = [3, -2]
y_points = [5, 0]

k = (y_points[1] - y_points[0]) / (x_points[1] - x_points[0])
m = y_points[0] - k * x_points[0]
a = k * x + m

print(k)

plt.scatter(x_points, y_points, color = "green")
plt.plot(x, y)
plt.plot(x, a)
plt.grid(True)
```
